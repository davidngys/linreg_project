## 2. Null Values <a class="anchor" id="chapter2"></a>
---
It is observed that there are multiple columns with missing values. A dataframe will be contructed to visualise the null values better.

```python
# Create null values dataframe
train_null_df = pd.DataFrame(train_df.isnull().sum(),columns= ['null']).sort_values(by='null',ascending=False)
# Insert column to see % of null values in that column
train_null_df['%_null'] = train_null_df['null']/2051*100
```

```python
# Filter columns with null values
display(train_null_df[train_null_df['null'] > 0])
```

### 2.1 Removing columns <a class="anchor" id="section_2_1"></a>
---
5 features have been identified to have >32% of null values and will be removed from the dataframe.
- Pool QC
- Misc Feature
- Alley
- Fence
- Fireplace Qu  

Rationale for using 32%:
- Normal distributions have 68% of the area within 1 standard deviation of the mean
- Alternative to removing these columns would be to fill the null values with the mean of that column
- As these columns have more than 32% of the data missing, filling with the mean values will not create a meaningful parameter for our model

In addition to columns removed due to null values, the following columns will also be removed:
- PID: Represents unique parcel identification number that has no relation to target (sales price)

```python
# Dropping columns and creating clean datafram object
clean_train_df = train_df.drop(['Pool QC','Misc Feature','Alley','Fence','Fireplace Qu','PID'],axis=1)
display(clean_train_df.head())
```

### 2.2 Filling NA values in remaining columns <a class="anchor" id="section_2_2"></a>
---
#### Lot Frontage  
Lot frontage refers to the linear distance of street connected to the property in feet. Based on the assumption that the lot frontage should be relatively similar for each property in a similar zone, we look further into statistics of lot frontage grouped by MS Zoning.

```python
display(clean_train_df[['Lot Frontage','MS Zoning']].groupby('MS Zoning').describe())
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x = 'MS Zoning', y='Lot Frontage', data = clean_train_df,orient="v",)
plt.tight_layout()
```

With the exception of FV (Floating Village Residential) and RH (Residential High density) zones, we see that the Lot Frontage values are relatively closely spread around the mean.

The Lot Frontage null values will be filled with the mean based on the MS Zoning.

```python
# Fill Lot Frontage null values with mean grouped by MS Zoning
clean_train_df['Lot Frontage'] = clean_train_df['Lot Frontage'].fillna(clean_train_df.groupby('MS Zoning')['Lot Frontage'].transform('mean'))
# Check that Lot Frontage now has no more null values.
display(clean_train_df['Lot Frontage'].isnull().sum())
```

Successfully filled all null values with mean based on MS Zoning.

#### Remaining columns with NA values
For the other remaining columns with NA values, as the % of NA values is low (< 6%), we will take the following approach.
- For numerical columns, replace with **mean**
- For non-numerical columns, replace with **mode**

```python
# Get numeric columns
clean_train_df_num = clean_train_df.select_dtypes(exclude='object')
# Get non-numeric columns
clean_train_df_obj = clean_train_df.select_dtypes(include='object')
```

```python
# Fill numeric columns with mean of that column
for col in clean_train_df_num.columns:
    clean_train_df[col].fillna(clean_train_df[col].mean(),inplace=True)
    
# Fill non-numeric columns with mode of that column
for col in clean_train_df_obj.columns:
    clean_train_df[col].fillna(clean_train_df[col].mode()[0],inplace=True)
    
# Check if null values have been successfully filled
clean_train_df.info()
```

```python
clean_train_df.head()
```

```python
# Checking that all null values have been imputed.
print(all(clean_train_df.isnull().sum() == 0))
```

## 3. Summary on Steps Taken <a class="anchor" id="chapter3"></a>

---

**Data reading and validation**
- Read in data from CSV file
- Checked that values in Ids and PIDs are unique to ensure no repeated values

**Cleaning of Null Data**
- Removed columns with >32% of null values in data (note that columns removed from this step will also be done for the test set regardless of number of null values in test set) 
- Replaced missing Lot Frontage values with mean values grouped by MS Zoning
- Replaced numeric columns with null values to mean of that column
- Replaced non-numeric columns with null values to mode of that column

These steps will be repeated for test data set.

## 4. Defining Functions for Data Cleaning <a class="anchor" id="chapter4"></a>
---

```python
def data_valid(df):
    # Checking if all rows within data set are unique
    if len(df['Id'].unique()) == len(df['Id']) and len(df['PID'].unique()) == len(df['Id']):
        print("There are no duplicates.")
    else:
        print("Please identify duplicates and remove them.")
    # Displaying info on data sets
    display(df.info())
    display(df.describe())
```

```python
def clean_null(df):
    # Remove columns defined in methodology above
    clean_df = df.drop(['Pool QC','Misc Feature','Alley','Fence','Fireplace Qu','PID'],axis=1)
    # Replace missing Lot Frontage values with mean values grouped by MS Zoning
    clean_df['Lot Frontage'] = clean_df['Lot Frontage'].fillna(clean_df.groupby('MS Zoning')['Lot Frontage'].transform('mean'))
    # Replace numeric columns with null values to mean of that column
    clean_df_num = clean_df.select_dtypes(exclude='object')
    for col in clean_df_num.columns:
        clean_df[col].fillna(clean_df[col].mean(),inplace=True)
    # Replace non-numeric columns with null values to mode of that column
    clean_df_obj = clean_df.select_dtypes(include='object')
    for col2 in clean_df_obj.columns:
        clean_df[col2].fillna(clean_df[col2].mode()[0],inplace=True)
    # Return cleaned dataframe
    return clean_df
    
```

## 5. Read In and Clean Test Data <a class="anchor" id="chapter5"></a>
---

```python
# Save test data set as object 'test_df'
test_df = pd.read_csv("../datasets/test.csv")
```

```python
# Run data validation function on test data set
data_valid(test_df)
```

```python
# Run cleaning data function on test data set
clean_test_df = clean_null(test_df)
```

```python
# Checking that there are no more null values in test data
clean_test_df.info()
```

```python
# Ensuring that parameters in cleaned training data and testing data are the same. 
# Returns 'True' if they are the same.
print(all(clean_train_df.drop('SalePrice',axis=1).columns == clean_test_df.columns))
```
